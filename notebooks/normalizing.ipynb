```python
import pandas as pd
import re

# Emoji pattern to match emojis in text
emoji_pattern = re.compile("["
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F1E0-\U0001F1FF"  # flags (iOS)
    "\U00002700-\U000027BF"  # dingbats
    "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
    "]+", flags=re.UNICODE)

# Function to remove emojis from a string
def remove_emojis(text):
    if isinstance(text, str):
        return emoji_pattern.sub(r'', text)
    return text

# Load your data
df = pd.read_csv('../data/telegram_data.csv')

# Remove emojis in 'Message' column
df['Message'] = df['Message'].apply(remove_emojis)

# (Optional) save cleaned data
df.to_csv('../data/cleaned_no_emoji.csv', index=False)
```

```python
import pandas as pd
import re

# Pattern to detect emojis
emoji_pattern = re.compile("["
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F1E0-\U0001F1FF"  # flags
    "\U00002700-\U000027BF"  # dingbats (includes some symbols like ✅)
    "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
    "]+", flags=re.UNICODE)
    
def remove_emojis(text):
    if isinstance(text, str):
        # Remove emojis, keep everything else intact
        return emoji_pattern.sub(r'', text)
    return text

# Load your data
df = pd.read_csv('../data/telegram_data(2).csv')

# Remove emojis from the 'Message' column
df['Message'] = df['Message'].apply(remove_emojis)

# (Optional) save cleaned data
df.to_csv('../data/cleaned_messages_demo1.csv', index=False, encoding='utf-8-sig')

```
